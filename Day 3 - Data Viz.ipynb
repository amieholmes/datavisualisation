```python
#import the real data
housing = pd.read_csv('boston_housing_data.csv')

drink_cols = ['country', 'beer', 'spirit', 'wine', 'litres', 'continent']
drinks = pd.read_csv('drinks.csv')
drinks.columns = drink_cols

ufo = pd.read_csv('ufo.csv')
```

```python
ufo
```

```python
# new column made by splitting time and date
ufo['Time'] = pd.to_datetime(ufo.Time) # clean the time column by making it into a DATETIME Object
ufo['Year'] = ufo.Time.dt.year # From the time column, extract year and put it in a new column called year
```

```python
drinks.continent.value_counts()
```

```python
#bar graphs for categorical data

drinks.continent.value_counts().plot(kind='bar');
```

```python
#clustered bar chart
drinks.groupby('continent').mean().plot(kind='bar');
```

```python
#sorted lowest to highest through beer servings
drinks.groupby('continent').mean().sort_values('beer').plot(kind='bar');
```

```python
#get rid of litres
drinks.groupby('continent').mean().drop('litres', axis=1).plot(kind='bar');
```

```python
#stacked bar chart
drinks.groupby('continent').mean().drop('litres', axis=1).plot(kind='bar', stacked=True);
```
